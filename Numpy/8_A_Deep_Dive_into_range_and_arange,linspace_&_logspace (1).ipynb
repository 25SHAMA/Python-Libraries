```python
import numpy as np
```

###1.Range

```python
r = range(10)
print(r)

r = range(0,10,2)
print(r)

r = range(10,0,-2)
print(r)

print(5 in r)
print(6 in range(10))
print(list(r))
print(list(range(10)))
```

###2.arange

```python
print(np.arange(0,10,2))
```

```python
print(np.arange(10,0,-1))
```

```python
print(np.arange(0,10,.5))#=>this we cannot do it in range function
```

###3.linspace==>equal intervals we create the numbers

```python
l = np.linspace(0,10,5)#it will give five equal steps between the given range
print(l)
```

```python
l = np.arange(0,16,5)
print(l)
```

```python
l = np.linspace(1,10,9)
print(l)
```

```python
l = np.linspace(1,10,9,endpoint= False)
print(l)
```

```python
l = np.linspace(1,10,9,retstep = True)#=>retstep = gives what step we are taking
print(l)
```

```python
l,s = np.linspace(1,10,9,retstep = True)#=>retstep = gives what step we are taking
print(l)
print(s)
```

###4.logspcae

```python
arr = np.logspace(0,10,5)
print(arr)
```

```python
arr = np.logspace(0,10,5,endpoint = False)
print(arr)
```

```python
arr = np.logspace(0,10,5,base = 2)
print(arr)
```
