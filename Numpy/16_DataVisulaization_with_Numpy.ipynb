```python
import numpy as np
import matplotlib.pyplot as plt
```

##1.Line Plot

##### 1.1)Basic Line Plot

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cosine Wave')
plt.show()
```

##### 1.2)Custom Line Graph

```python
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y,'o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cosine Wave')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y,'--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cosine Wave')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y,'o',linestyle='--',color='r',markersize = 5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cosine Wave')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y,'o',linestyle='--',color='r',markersize = 5,alpha = .5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cosine Wave')
plt.show()
```

##### 1.3)Line plot with multiple lines

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,'o',linestyle='--',color='r',markersize = 5,alpha = .5,label = 'sin(x)')
plt.plot(x,y2,'o',linestyle='--',color='b',markersize = 5,alpha = .5,label = 'cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()
plt.show()
```

##2.Scatter Plot

##### 2.1)Basic Scatter Plot

```python
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

##### 2.2)Customized Scatter Plot

```python
x = np.random.rand(300)
y = np.random.rand(300)
z = np.random.rand(300)

plt.scatter(x,y,marker = 'o',c = z,alpha = .8,cmap = 'viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.colorbar()
plt.show()
```

##### 2.3)Scatter plot with multiple classes

```python
x1 = np.random.rand(200)
y1 = np.random.rand(200)

x2 = np.random.rand(200)
y2 = np.random.rand(200)

plt.scatter(x1,y1,marker = 'o',alpha = .7)
plt.scatter(x2,y2,marker = 'o',alpha = .7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.colorbar()
plt.show()
```

```python
x1 = np.random.rand(200)
y1 = np.random.rand(200)

x2 = np.random.rand(200)
y2 = np.random.rand(200) + 1 #=>the points will be shifted to one side

plt.scatter(x1,y1,marker = 'o',alpha = .7)
plt.scatter(x2,y2,marker = 'o',alpha = .7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.colorbar()
plt.show()
```

```python
x1 = np.random.rand(200)
y1 = np.random.rand(200)

x2 = np.random.rand(200)
y2 = np.random.rand(200) + 0.5 #=>half of them will shift

plt.scatter(x1,y1,marker = 'o',alpha = .7)
plt.scatter(x2,y2,marker = 'o',alpha = .7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.colorbar()
plt.show()
```

```python
x1 = np.random.rand(200)
y1 = np.random.rand(200)

x2 = np.random.rand(200) + 0.5 #=>half of them will shift
y2 = np.random.rand(200) + 0.5

plt.scatter(x1,y1,marker = 'o',alpha = .7)
plt.scatter(x2,y2,marker = 'o',alpha = .7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.colorbar()
plt.show()
```

```python
x1 = np.random.rand(200) - 0.5
y1 = np.random.rand(200) - 0.5

x2 = np.random.rand(200) + 0.5
y2 = np.random.rand(200) + 0.5

plt.scatter(x1,y1,marker = 'o',alpha = .7)
plt.scatter(x2,y2,marker = 'o',alpha = .7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.colorbar()
plt.show()
```

## 3.Bar Graph

##### 1.1)Basic Bar Graph

```python
x = np.array(['A','B','C','D'])
y = np.array([10,20,30,40])
plt.bar(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph')
plt.show()
```

##### 1.2)Staked Bar Graph

```python
data = np.random.rand(10,4)
plt.bar(np.arange(10),data[:,0] + data[:,1] + data[:,2] + data[:,3],label = 'Dimension-4')
plt.bar(np.arange(10),data[:,0] + data[:,1] + data[:,2] , label = 'Dimension-3')
plt.bar(np.arange(10),data[:,0] + data[:,1],label = 'Dimension-2')
plt.bar(np.arange(10),data[:,0],label = 'Dimension-1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph')
plt.legend()
plt.show()
```

##### 3.3)Grouped Bar Graph

```python
data = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,90]])
width = 0.25
x = np.arange(3)

plt.bar(x - width,data[:,0],label = 'Group-1',width = .25)
plt.bar(x,        data[:,1],label = 'Group-2',width = .25)
plt.bar(x + width,data[:,2],label = 'Group-3',width= .25)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

##4.Histogram

##### 4.1)Basic Histogram

```python
data = np.random.randn(1000)
plt.hist(data)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

```python
data = np.random.randn(10000000)
plt.hist(data,bins = 1000)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

##### 4.2)Accumulative Histogram

```python
data = np.random.randn(1000)
plt.hist(data,bins = 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data,bins = 10,cumulative = True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.show()
```

#####4.3)Histogram With Multiple Datasets

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)

plt.hist(data1,bins = 10,alpha = .5,label = 'Data-1')
plt.hist(data2,bins = 10,alpha = .5,label = 'Data-2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.legend()
plt.show()
```

```python
data1 = np.random.randn(10000)
data2 = np.random.randn(10000) + 3 #=>this category shifts to rightside because we have +3 to the data

plt.hist(data1,bins = 30,alpha = .5,label = 'Data-1')
plt.hist(data2,bins = 30,alpha = .5,label = 'Data-2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.legend()
plt.show()
```

```python
data1 = np.random.randn(10000)
data2 = np.random.randn(10000) + 6

plt.hist(data1,bins = 40,alpha = .5,label = 'Data-1')
plt.hist(data2,bins = 40,alpha = .5,label = 'Data-2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram')
plt.legend()
plt.show()
```

## 7.Pie Chart

```python
labels = ['Python','C++','Java','JavaScript','PHP']
sizes = [21.3,14.6,9.2,8.5,7.7]
explode = (0.13,.1,0,0,0)
colors = ['gold','yellowgreen','lightcoral','lightskyblue','lightpink']

plt.title('Programming Language Popularity')
plt.pie(sizes,labels = labels,explode = explode,shadow = True,colors = colors,autopct = '%1.2f%%',startangle=140)
plt.axis('equal')
plt.legend(title = 'Languages',loc ='upper right')
plt.show()
```
