###1.Arrays vs Lists Comparision

```python
lst = [1,2,3]
print(lst)
print(type(lst))
```

```python
import numpy as np
arr = np.array([1,2,3])
print(arr)
print(type(arr))
```

```python
#comparision
print(arr == lst)
```

```python
print(arr.tolist() == lst)#array converted to list and compared list with list so both are equal it returned true
```

```python
lst = [1,3,2]
arr = np.array([1,2,3])

print(lst == arr)
print(sorted(lst) == sorted(arr))
```

###2.Addding elements in an Array vs List

```python
lst = [1,3,2]
arr = np.array([1,2,3])
print(lst)
lst.append(2)
print(lst)

print('-----'*5)
print(arr)
arr = np.append(arr,4)
print(arr)
```

###3.Removing elements in an Array vs List

```python
lst = [1,3,2]
arr = np.array([1,2,3])
print(lst)
lst.pop() #or del lst[-1]--->deleting the last element-using index
print(lst)

print('-----'*5)
print(arr)
arr = np.delete(arr,-2)
print(arr)
```

###4.Updating elements in an Array vs List

```python
lst = [1,3,2]
arr = np.array([1,2,3])

print(lst)
lst[-1] = 5
lst[-3] = 6
print(lst)

print('---'*5)
print(arr)
arr[-1] = 5
arr[-3] = 6
print(arr)
```

###5.Comparision between Array vs List

```python
import time

#creating the data
my_list = list(range(1000000))
my_array = np.arange(1000000)

#search time for list
start_time = time.time()
for i in range(1000):
  if 5000000 in my_list:
    pass
end_time = time.time()
print('List :' ,end_time - start_time)


#search time for array
start_time = time.time()
for i in range(1000):
  if 5000000 in my_array:
    pass
end_time = time.time()
print('Array :',end_time - start_time)
```
