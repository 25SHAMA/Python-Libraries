```python
import numpy as np
```

###1.Airthmetic

```python
a = np.array([1,2,3])
b = np.array([[10,20,30],[40,50,60]])

print(a)
print(b)

#Addition
print('-'*20)
print(a+5)
print('*'*10)
print(a+b)

#Subtraction
print('-'*20)
print(a-2)
print('*'*10)
print(a-b)
print('*'*10)
print(b-a)

#Multiplication
print('-'*20)
print(a*2)
print('*'*10)
print(a*b)

#Division
print('-'*20)
print(a /2)
print('*'*10)
print(a/b)

#Exponential
print('-'*20)
print(a**2)
print('*'*10)
print(a**b)

#Modulo
print('-'*20)
print(a%2)
print('*'*10)
print(a%b)
```

###2.Comparision

```python
a = np.array([1,2,3])
b = np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
print('-'*20)
#>
print(a>b)
print('-'*20)
#<
print(a<b)
print('-'*20)
#>=
print(a>=b)
print('-'*20)
#<=
print(a<=b)
print('-'*20)
#!=
print(a!=b)
print('-'*20)
#==
print(a==b)
```

###3.Logical

```python
a = np.array([1,2,3])
b = np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
print('-'*20)
print(a & b)
print('-'*20)
print(a | b)
print('-'*20)
print(~a)
print('-'*20)
print(~b)
print('-'*20)
print(a ^ b)
```

###4.Bitwise

```python
a = np.array([1,2,3])
b = np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
print('-'*20)
print(np.bitwise_and(a,b))
print('-'*20)
print(np.bitwise_or(a,b))
print('-'*20)
print(np.bitwise_xor(a,b))
print('-'*20)
print(np.bitwise_not(a,b))
print('-'*20)
print(np.bitwise_left_shift(a,b))
print('-'*20)
print(np.bitwise_right_shift(a,b))
```
