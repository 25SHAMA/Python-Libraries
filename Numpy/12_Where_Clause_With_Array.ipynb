```python
import numpy as np
```

### 1D array

```python
np.random.seed(10)
arr = np.random.randint(1,11,size = 10)
print(arr)

print('*'*30)
print(np.where(arr>5))#gives indiced which satisfy the condition =>0-10,4-10,7-9,8-10,===>indices-Element
print('*'*30)
print(np.where(arr>=5))
print(arr[np.where(arr>=5)])#gives elements in an array
print('*'*30)
```

### 2D Array

```python
np.random.seed(10)
arr = np.random.randint(1,11,size = (3,3))
print(arr)

print('*'*30)
indices = np.where(arr>=5)#pairs ==> (0,0),(0,1),(1,1),(2,1),(2,2)
print(indices)
print('*'*30)
print(arr[indices])
```

###3D Array

```python
np.random.seed(10)
arr = np.random.randint(1,11,size = (3,3,3))
print(arr)

print('*'*30)
indices = np.where(arr>=5)#pairs ==> (0,0,0),(0,0,1),(0,1,1),(0,2,1),(0,2,2),(1,0,1)...etc
print(indices)
print('*'*30)
print(arr[indices])
```
