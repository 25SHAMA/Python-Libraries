###1.Converting DataStrutures to Numpy Arrays

```python
import numpy as np
```

**1.1)Converting a tuple to an array**

```python
tup = (1,2,3,4,5)
print(tup)
print(type(tup))

arr = np.array(tup)
print(arr)
print(type(arr))
```

**1.2)Converting a list to an array**

```python
lst = [1,2,3,4,5]
print(lst)
print(type(lst))

arr = np.array(lst)
print(arr)
print(type(arr))
```

**1.3)Converting a Dictionary to an array**

```python
dictionary = {1:'apple',2:'ball',3:'cat'}
print(dictionary)
print(type(dictionary))

arr = np.array(dictionary)
print(arr)
print(type(arr))

print('--'*20)
#Efficient way of converting dictionary into an array is adding key and value pairs separately
arr_keys = np.array(list(dictionary.keys()))
arr_values = np.array(list(dictionary.values()))
print(arr_keys)
print(type(arr_keys))
print(arr_values)
print(type(arr_values))
```

**1.4)Converting a set to an array**

```python
set_ = {1,2,3,4,5}
print(set_)
print(type(set_))

arr = np.array(set_)
print(arr)
print(type(arr))
```

**1.5)Converting String to array**

```python
#Two ways
arr = np.array('hello world')
print(arr)
print(type(arr))
arr = np.array(list('hello world'))
print(arr)
print(type(arr))
```

###2.Converting the Datatype of Numpy Array

**2.1)Create an array of integers**

```python
arr = np.array([1,2,3,4,5])
print(arr)
print(arr.dtype)

arr = arr.astype('float')
print(arr)
print(arr.dtype)

arr = arr.astype('str')
print(arr)
print(arr.dtype)

arr = arr.astype('bool')
print(arr)
print(arr.dtype)
```

**2.2)Create an array of floats**

```python
arr = np.array([1.,2.,3.,4.,5.])
print(arr)
print(arr.dtype)

arr = arr.astype('int')
print(arr)
print(arr.dtype)

arr = arr.astype('str')
print(arr)
print(arr.dtype)

arr = arr.astype('bool')
print(arr)
print(arr.dtype)
```

**2.3)Create an array of strings**

```python
arr = np.array(['1','2','3','4','5'])
print(arr)
print(type(arr))

arr_int = arr.astype('int')
print(arr_int)
print(arr_int.dtype)

arr_float = arr.astype('float')
print(arr_float)
print(arr_float.dtype)

arr_bool = arr.astype('bool')
print(arr_bool)
print(arr_bool.dtype)
#we caannot convert words such 'A' or 'Sam' in to int,float,bool only numbers we can convert
```

**2.4)Create an array of boolean**

```python
arr = np.array([True,False,False,False,True])
print(arr)
print(type(arr))

arr_int = arr.astype('int')
print(arr_int)
print(arr_int.dtype)

arr_float = arr.astype('float')
print(arr_float)
print(arr_float.dtype)

arr_string = arr.astype('str')
print(arr_string)
print(arr_string.dtype)
```
