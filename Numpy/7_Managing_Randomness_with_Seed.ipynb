```python
import numpy as np
```

```python
print(np.random.rand())
print('-'*20)
print(np.random.rand(3))
print('-'*20)
print(np.random.randint(10))
print('-'*20)
print(np.random.randint(10, size=(3,5)))
```

```python
#To get / remain the random value as same we use seed --> we wil not get everytime new value if we use seed
np.random.seed(2)

print(np.random.rand())
print('-'*20)
print(np.random.rand(3))
print('-'*20)
print(np.random.randint(10))
print('-'*20)
print(np.random.randint(10, size=(3,5)))
```

```python
np.random.seed(2)

print(np.random.rand())
print('-'*20)
print(np.random.rand(3))
print('-'*20)
print(np.random.randint(10))
print('-'*20)
print(np.random.randint(10, size=(3,5)))#=>we can see that for above code and now for the code we get the same value as before because we have used seed
```

```python
np.random.seed(3)

arr = np.random.randint(2, size = 1000)

print('Heads:',np.sum(arr == 1))
print('Tails:',np.sum(arr == 0))
```

```python
np.random.seed(5)
for i in range(10):
    arr = np.random.randint(2, size = 1000)

    print('Heads:',np.sum(arr == 1) ,"==>", 'Tails:',np.sum(arr == 0)) # for ten iterations we get like this
```
