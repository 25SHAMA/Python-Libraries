```python
import numpy as np
```

###1.Saving data in Files

- TextFile

```python
arr = np.array([[1,2,3],[4,5,6]])
np.savetxt('textfile.txt',arr,delimiter=',')
```

- npyFile

```python
arr = np.array([[1,2,3],[4,5,6]])
np.save('data.npy',arr)#Best way to save the File
```

- csvFile

```python
arr = np.array([[1,2,3],[4,5,6]])
np.savetxt('data.csv',arr,header = 'a,b,c')
```

- jsonFile

```python
arr = np.array([[1,2,3],[4,5,6]])
np.savetxt('file.json',arr,delimiter=',',header = '{"data" : [',footer = ']}')
```

- yaml

```python
arr = np.array([[1,2,3],[4,5,6]])
np.savetxt('file.yaml',arr,delimiter=',',header = '- data : ')
```

###2.Reading Data From Files

- TextFile

```python
data = np.loadtxt('textfile.txt',delimiter=',')
print(data)
```

- npyFile

```python
data = np.load('data.npy')#Best Way
print(data)
```

- csvFile

```python
data = np.loadtxt('data.csv')
print(data)
```

- jsonFile

```python
data = np.loadtxt('file.json',delimiter=',')
print(data)
```

- yamlFile

```python
data = np.loadtxt('file.yaml',delimiter=',')
print(data)
```

###3.Writing data in Files

###1.Replacing exisiting Content

```python
data = np.array([4,5,6])
np.savetxt('textfile.txt',data,delimiter=',')#Replaces the values (1,2,3,4,5,6) with 4,5,6
```

###2.Appending the data on existing File

```python
data = np.load('data.npy')

data = np.append(data,[[7,8,9]],axis=0)
np.save('data.npy',data)
```
